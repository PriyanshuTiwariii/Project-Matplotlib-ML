```python
import matplotlib.pyplot as plt

x = ["C","C++","JAVA","Python"]
y = [20, 25, 35, 55]

plt.plot(x, y, color='green', linewidth=3, marker='o',
         markersize=15, linestyle='--')


plt.title("Line Chart")

plt.ylabel("user's %")

plt.xlabel('LAUNGUAGE')

plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["C","C++","JAVA","Python"]
y = [20, 25, 35, 55]
c=["r","g","b","m"]

plt.bar(x, y, color=c, linewidth=3,)


plt.title("Line Chart")

plt.ylabel("user's %")

plt.xlabel('LAUNGUAGE')

plt.show()
```
